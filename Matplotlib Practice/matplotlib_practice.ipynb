```python
# Import dependencies
# Same as typing from matplotlib import pyplot
# pyplot module allows us to create all necessary charts - 
# create a plotting area in a figure, plot lines in plotting area, annotate chart w/ labels, colors, & other features
import matplotlib.pyplot as plt
```

```python
# Example #1 - Line Charts
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Example 1.0
# Create a Line Chart Using the MATLAB Method

# Create the chart
# plt.plot() creates the figure with the current axes
plt.plot(x_axis, y_axis)
```

```python
# Example 1.1.A
# Create a Line Chart Using the Object-Oriented Interface
# Create the plot with ax.plot() instead of plt.plot()

# Create a figure
fig, ax = plt.subplots()

# Define the axes
ax.plot(x_axis, y_axis)
```

```python
# Example 1.1.B
# Alternative way to create the plot with ax.plot()

# fig references the figure
# Use fig = plt.figure() to change figure-level attributes (axis labels, title, legend) or save figure as image
fig = plt.figure()

# ax references the axes of the figure and allows us to annotate the figure
ax = fig.add_subplot()

# plot the chart
ax.plot(x_axis, y_axis)
```

```python
# Example 1.1.C
# Create a quick graph using the object-oriented interface method

# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Example 1.2
# Demonstrate the plt.show() function
# plt.show() function looks for all active figure objects 
# and opens a window that displays them if you have >= 2 sets of data to plot

# Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Example 1.3.A
# Annotate Charts Using the MATLAB Method

# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label='Boston')
# Create the labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# Set the y limit between 0 and 45
plt.ylim(0, 45)
# Create a title 
plt.title("PyBer Fare by Month")
# Add the legend
plt.legend()
```

```python
# Example 1.3.B
# Amend the code block to add:
# new line color, modified line width, and a marker for data points

# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create the labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# Set the y limit between 0 and 45
plt.ylim(0, 45)
# Create a title 
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add the legend
plt.legend()
```

```python
# Example 1.3.C
# Use xlim() to show the months January - June

# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create the labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# Set the y limit between 0 and 45
plt.ylim(0, 45)
# Set the x limit between Jan. and June
#plt.xlim("Jan", "June")
# Set the x limit between Jan. and June: Alternate
plt.xlim(0, 5)
# Create a title 
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add the legend
plt.legend()
```

```python
# Example 1.3.D
# Change the figure size to 20 pixels wide by 10 pixels high

# Change the figure size
#plt.figure(figsize=(20, 10))
# Change the figure size: Alternative
plt.subplots(figsize=(20, 10))
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create the labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# Set the y limit between 0 and 45
plt.ylim(0, 45)
# Create a title 
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add the legend
plt.legend()
```

```python
# Example 1.4.A
# Annotate Charts Using the Object-Oriented Interface

# Create a figure
fig, ax = plt.subplots()

# Define the axes 
# Change the line color to green
# Change the marker to a diamond
# Change the line width to 2 points
# Add a label that will be added to the legend
ax.plot(x_axis, y_axis, color='green', marker='D', linewidth=2.0, label='Boston')

# Add labels to the axes
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')

# Set the min and max range of the y-axis
ax.set_ylim(0, 45)

# Add a title
ax.set_title("PyBer Fare by Month")

# Add a grid to the chart
ax.grid()

# Add the legend
ax.legend()
```

```python
# Example 3.4.B
# Annotate the Bubble Chart

# Create the plot
# Switch the axis so that Fare ($) data is on the x-axis
# Change the color of the markers to sky blue
# Change the size of the markers to 5x each data point
# Make the color 20% transparent
# Add a black edge color to the circles 
# Add a legend for the city of Boston
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c='xkcd:sky blue', s= [i * 5 for i in y_axis], alpha=.8, edgecolors='black',linewidths=2, label='Boston')
ax.legend()

# Add a title
ax.set_title("PyBer Fare by Month")

# Add axes labels
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")

# Add a limit for the x-axis data
ax.set_xlim(0, 50)

# Invert the y-axis so that January is at the top
ax.invert_yaxis()
```

```python
# Example #4 - Pie Charts
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Example 4.1.A
# Create a Pie Chart Using the MATLAB Method

# Use the x_axis (months) as labels, and y_axis (total fare per month) as values
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Example 4.1.B
# Create a Pie Chart Using the MATLAB Method

# The largest % is July, the 7th value in the x_axis
explode_values = [0,0,0,0,0,0,0.2,0,0,0,0,0]
# Add percentages for each month and "explode" the largest percentage
plt.pie(y_axis, labels=x_axis, explode=explode_values, autopct='%.1f%%')
plt.show()
```

```python
# Example 4.1.C
# Create a Pie Chart Using the MATLAB Method

# The largest % is July, the 7th value in the x_axis array
explode_values = [0,0,0,0,0,0,0.2,0,0,0,0,0]
# Increase the size of the chart
plt.subplots(figsize=(8, 8))
# Add percentages for each month and "explode" the largest percentage
plt.pie(y_axis, labels=x_axis, explode=explode_values, autopct='%.1f%%')
plt.show()
```

```python
# Example 4.1.D
# Create a Pie Chart Using the MATLAB Method

# Assign 12 colors, one for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
# "explode" the largest percentage
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
# Increase the chart size
plt.subplots(figsize=(8, 8))
# Plot the chart
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Example 4.2.A
# Create a Pie Chart Using the Object-Oriented Interface

fig, ax = plt.subplots()
ax.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Example 4.2.B
# Create a Pie Chart Using the Object-Oriented Interface

# Explode the two highest percentage months
explode = [0,0,0.2,0,0,0,0.2,0,0,0,0,0]

# Add new colors 
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

# Increase the figure size to 8x8
fig, ax = plt.subplots(figsize=(8,8))

# Add a percentage to one decimal place to each wedge of the pie
# "explode" the values and include the new colors
# Add a shadow
# Add a start angle so that January is at the top
# Reverse the order so that the month order is in a clockwise direction
ax.pie(y_axis, labels=x_axis, autopct="%.1f%%", explode=explode, colors=colors, shadow=True, startangle=95, counterclock=False)

plt.show()
```
