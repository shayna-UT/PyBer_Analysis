```python
# Bar Charts:
# Good at displaying discrete data (i.e., data based on counts) in distinct columns
# Clearly compare two or more datasets (display them side by side or stacked)
# More visual appeal and info compared to having multiple lines on the same graph, 
# which can mask the underlying information
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the y-axis equal to a list of savings
savings = [0,100,150,200,300,500,300,350,500,600,750,1000]
# Set the x-axis equal to a list of strings for each month
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
# Vertical Bar Chart:
# use for graphing time-series data or ordinal values (age ranges, salary ranges, groups)

# Create the plot
plt.bar(months, savings)
plt.xlabel('Month')
plt.ylabel('Savings ($)')
plt.title('Savings Account Results')
plt.show()
```

```python
# Horizontal Bar Chart:
# use for graphing nominal variables (favorite music groups among teens, # of votes by county or state)
# where you sort the data from greatest to least or least to greatest 

# Create the plot
plt.barh(months, savings)
plt.xlabel('Savings ($)')
plt.ylabel('Month')
plt.title('Savings Account Results')
plt.show()
```
