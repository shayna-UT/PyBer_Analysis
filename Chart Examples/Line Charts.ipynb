```python
# Line Charts: 
# plot continuous data as points and joins them w/ a line
# allow you to compare multiple datasets on the same chart easily
# diplay data over time (relationship btw. time & another value)
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the y-axis equal to a list of savings
savings = [0,100,150,200,300,500,300,350,500,600,750,1000]
# Set the x-axis equal to a list of strings for each month
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
# Create the plot
plt.plot(months, savings)
plt.xlabel('Month')
plt.ylabel('Savings ($)')
plt.title('Savings Account Results')
plt.show()
```
