```python
# Pie Charts:
# Way to visualize percentages of categories as wedges of a pie
# Visualize a part of a whole
# limitation - pie charts can only represent one dataset / category 
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Variables
categories = ['pants', 'shirts', 'outerwear', 'pajamas', 'underwear', 'shoes']
revenue = [25, 15, 5, 20, 10, 25]
```

```python
# Plot the pie chart
plt.pie(revenue)
plt.show()
```

```python
# Label the chart
plt.pie(revenue, labels=categories)
plt.show()
```

```python
# Add the percentages into the pie chart
plt.pie(revenue, labels=categories, autopct="%1.1f%%")
plt.show()
```

```python
# Highlight a specific category (explode() argument)

# the "explode" parameter will offset the indicated wedge by a fraction of the radius, 
# where "0" is zero distance fromt the center of the pie
# and "1" is completely outside the diameter of the pie 
explode = [0,0.1,0,0,0,0]
plt.pie(revenue, labels=categories, autopct="%1.1f%%", explode=explode)
plt.show()
```
