```python
# Scatter Plots:
# made up of multiple data points across the x- and y- plane and show relationship btw. 2 variables
# visually depicts a positive, negative, or no-correlation relationship btw. 2 factors
# used when developing regression models to yield a line of best fit & used to make predictions for y variables when given x
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Independent variable (x) - "Total # of Rides"
x = [50, 42, 39, 34, 32, 48, 22, 17, 8, 13]
# Dependent variable (y) - "Average Fare"
y = [17, 19, 21, 22, 24, 23, 34, 31, 43, 35]
```

```python
# Create a scatter plot
# Set the color to green
# Change the shape of the markers
# Change the size of the markers
plt.scatter(x,y, c='g', marker='x', s=100)
# Add a title and axis labels
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Average Fare vs. Total Rides")
# Show the plot
plt.show()

# Analysis: you can there is some negative relationship btw. total rides and avg. fare
```

```python
# Create a scatter plot
# Set the color to green
# Change the shape of the markers
# Change the size of the markers to be dependent on the total # of rides
plt.scatter(x,y, c='g', marker='x', s=x)
# Add a title and axis labels
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Average Fare vs. Total Rides")
# Show the plot
plt.show()

# Analysis: you can there is some negative relationship btw. total rides and avg. fare
```
